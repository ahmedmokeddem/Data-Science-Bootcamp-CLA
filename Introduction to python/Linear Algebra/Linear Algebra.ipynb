```python
from matplotlib import image
from matplotlib import pyplot
import numpy as np
```

```python
def matrix_transpose(array,axes=(1,0,2)): 
    return np.transpose(array,axes)
```

```python
apple=image.imread("apple.jpg")
Tapple=matrix_transpose(apple)
pyplot.imshow(Tapple)
pyplot.show()
```

```python
#Check if the matrix is symetric

def is_symetric(matrix):
    return np.all(matrix==matrix.T)
```

```python
#let's take the red matrix from this RGB image 
red_apple=apple[:,:,0]
red_apple
```

```python
pyplot.imshow(red_apple)
pyplot.show()
```

```python
is_symetric(red_apple)
```

```python
#Check if the matrix is triagonal
def is_triagonal(matrix):
    return np.all(matrix==np.triu(matrix)) or np.all(matrix==np.tril(matrix)) #upper or lower
```

```python
is_triagonal(red_apple)
```

```python
#Check if the matrix is diagonal
def is_diagonal(matrix):
    return np.all(matrix==np.diag(np.diag(matrix)))
```

```python
is_diagonal(red_apple)
```

```python
# check if the matrix equals identity matrix
def is_identity(matrix):
    return np.all(matrix==np.eye(matrix.shape[0]))
```

```python
is_identity(red_apple)
```
