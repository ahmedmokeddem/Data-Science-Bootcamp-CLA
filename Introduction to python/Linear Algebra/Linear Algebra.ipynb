```python
from matplotlib import image
from matplotlib import pyplot
import numpy as np
```

```python
def matrix_transpose(array,axes=(1,0,2)): 
    return np.transpose(array,axes)
```

```python
apple=image.imread("apple.jpg")
Tapple=matrix_transpose(apple)
pyplot.imshow(Tapple)
pyplot.show()
```

```python
#Check if the matrix is symetric

def is_symetric(matrix):
    return np.all(matrix==matrix_transpose(matrix))
```

```python
is_symetric(apple)
```

```python
#Check if the matrix is triagonal
def is_triagonal(matrix):
    return np.all(matrix==np.triu(matrix)) or np.all(matrix==np.tril(matrix)) #upper or lower
```

```python
is_triagonal(apple)
```

```python
#Check if the matrix is diagonal
def is_diagonal(matrix):
    print(np.diagonal(matrix))
    #return np.all(matrix==np.diag(matrix))
```
